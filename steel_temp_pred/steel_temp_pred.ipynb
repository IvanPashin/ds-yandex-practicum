### Описание этапа обработки

Сталь обрабатывают в металлическом ковше вместимостью около 100 тонн. Чтобы ковш выдерживал высокие температуры, изнутри его облицовывают огнеупорным кирпичом. Расплавленную сталь заливают в ковш и подогревают до нужной температуры графитовыми электродами. Они установлены в крышке ковша. 


Из сплава выводится сера (десульфурация), добавлением примесей корректируется химический состав и отбираются пробы. Сталь легируют — изменяют её состав — подавая куски сплава из бункера для сыпучих материалов или проволоку через специальный трайб-аппарат (англ. tribe, «масса»).


Перед тем как первый раз ввести легирующие добавки, измеряют температуру стали и производят её химический анализ. Потом температуру на несколько минут повышают, добавляют легирующие материалы и продувают сплав инертным газом. Затем его перемешивают и снова проводят измерения. Такой цикл повторяется до достижения целевого химического состава и оптимальной температуры плавки.


Тогда расплавленная сталь отправляется на доводку металла или поступает в машину непрерывной разливки. Оттуда готовый продукт выходит в виде заготовок-слябов (англ. slab, «плита»).

Есть предположение, что начиная с 2500 партии датчик температуры замеряет только первое значение, проверим его

```python
def sensor_check(steel_batch):
    # проверим первую запись перед нагревом
    first_is_correct = not np.isnan(steel_batch.iloc[0])
    # остальные замеры не произведены
    other_is_nan = np.isnan(steel_batch.iloc[1:]).all()
    return first_is_correct and other_is_nan
```

```python
temp.query('key in @keys_temp_nan').groupby(['key'])['temp'].agg(sensor_check).all()
```

Предположение подтвердилось, пропущенные значения целевой переменной восстановить нам не удастся и придется удалить эти партии

```python
temp = temp.query('key not in @keys_temp_nan')
```

Так как мы предсказываем последнее значение температуры, то в выборку должы войти только те партии, в которых есть не менее двух измерений температуры

```python
# партии у которых один замер температуры
one_measurement = temp['key'].value_counts()[temp['key'].value_counts() == 1]
one_measurement
```

```python
temp = temp.query('key not in @one_measurement.index')
```

Пусть признаками у нас будут первый замер температуры, время между первый и последним замером, количество энергии и количество вспомогательных веществ. На повышение температуры влияет переданная электродами энергия. На понижение - добавление материалов и обдув, а также длительнось времени всего процесса (разница между последним и первым замером температуры).

```python
# Оставим номер партии, величину и время первого и последнего замера
temp = temp.groupby(["key"], as_index=False).agg(
    {"temp": ["first", "last"], "temp_time": ["first", "last"]}
)

# уберем мультиинтекс у названий столбцов
temp.columns = ["_".join(col) for col in temp.columns.values]
temp = temp.rename(columns={"key_": "key"})
```

```python
# посчитаем время всего процесса (в минутах)
temp["duration"] = (temp["temp_time_last"] - temp["temp_time_first"]) / pd.Timedelta(minutes=1)

# удалим столбцы с датами
temp = temp.drop(columns=['temp_time_last', 'temp_time_first'])
```

Вероятно, это ошибка датчика или переноса данных. Предлагаю заменить аномальное значение, с помощью расчета через известные даннве для этой партии. 

У потребителей электроэнергии значение коэфициента мощности (безразмерная физическая величина, характеризующая потребителя переменного электрического тока с точки зрения наличия в нагрузке реактивной составляющей) находится в районе 0.7-0.8.



```python
# коэффициент мощности для известных значений 2116 партии
arc_2116 = arc.query("key == 2116 and r_pwr > 0")
arc_2116["a_pwr"] / np.sqrt(arc_2116["a_pwr"] ** 2 + arc_2116["r_pwr"] ** 2)
```

```python
# среднее значение коэффициента мощности для 2116 партии
mean_power_factor_2116 = (
    arc_2116["a_pwr"] / np.sqrt(arc_2116["a_pwr"] ** 2 + arc_2116["r_pwr"] ** 2)
).mean()
mean_power_factor_2116
```

```python
# расчет рективной мощности через средний коэффицие
arc.loc[9780, "r_pwr"] = np.sqrt(
    (arc.loc[9780, "a_pwr"] / mean_power_factor_2116) ** 2
    - (arc.loc[9780, "a_pwr"]) ** 2
)
```

```python
arc.query("key == 2116")
```

Непосредсвенно на температуру вещества влияет переданная ему энергия (поизведение мощности на время передачи). При этом известно, что реактивная мощность не учавствует в нагреве электродов, и потребители обычно не платят за нее (в задании нет указаний на обратное). Введем новые признаки:
 
 - `heating_time` — время нагрева (разница между `arc_off` и `arc_on`), понадобится для расчета потраченной активной энергии
 - `a_energy` — потраченную активную энергию на нагрев (произведений `a_pwr` и `heating_time`)
 - `power_factor` — коэффицие́нт моощности (безразмерная физическая величина, характеризующая потребителя переменного электрического тока с точки зрения наличия в нагрузке реактивной составляющей). Расчитывается как отношение активной мощности к полной.

```python
arc["heating_time"] = (arc["arc_off"] - arc["arc_on"]) / pd.Timedelta(seconds=1)
arc['a_energy'] = arc['a_pwr'] * arc["heating_time"]
arc["power_factor"] = arc["a_pwr"] / np.sqrt(arc["a_pwr"] ** 2 + arc["r_pwr"] ** 2)
```

### Обучение моделей

Мы решаем задачу регрессии, в качестве моделей обучения будем использовать `LinearRegression`, `Ridge`, `Lasso`, `CatBoostRegressor`, `LGBMRegressor` и `RandomForestRegressor`. Подбирать гиперпараметры классификаторов с наилучшей метрикой качества *MAE* будем с помощью `RandomSearchCV`.

В датафрейме все признаки количественные, к ним будем применять масштабирование (StandardScaler и MinMaxScaler), если того требует соответствующий алгоритм обучения.

```python
num_features = train.drop(columns=['temp_last']).columns.to_list()
num_features
```

```python
ct = ColumnTransformer(
    [('scaler', StandardScaler(), num_features)])

pipe = Pipeline([('preprocessor', ct), 
                 ('regressor', Ridge())])
```

```python
# создание сетки гиперпараметров
parameters = [
    {
        'preprocessor__scaler': ['passthrough'],
        'regressor': [DummyRegressor(strategy='mean')],
    },
    {
        "preprocessor__scaler": [StandardScaler(), MinMaxScaler()],
        "regressor": [LinearRegression()],
    },
    {
        "preprocessor__scaler": [StandardScaler(), MinMaxScaler()],
        "regressor": [Ridge()],
        "regressor__alpha": np.logspace(-7, 1, 10),
    },
    {
        "preprocessor__scaler": [StandardScaler(), MinMaxScaler()],
        "regressor": [Lasso()],
        "regressor__alpha": np.logspace(-7, 1, 10),
    },
    {
        "regressor": [CatBoostRegressor(verbose=0, random_state=RANDOM_STATE)],
        "regressor__n_estimators": range(100, 500, 100),
        "regressor__max_depth": [4, 5, 6],
        "regressor__learning_rate": [0.003, 0.03, 0.1],
        "regressor__l2_leaf_reg": np.concatenate(
            (np.logspace(-20, -2, 10), np.linspace(1, 10, 10))
        ),
        "preprocessor__scaler": ["passthrough"],
    },
    {
        "regressor": [LGBMRegressor(random_state=RANDOM_STATE)],
        "regressor__max_depth": [-1, 3, 5, 7],
        "regressor__n_estimators": [50, 100, 200, 500],
        "regressor__learning_rate": [0.1, 0.03, 0.003],
        "preprocessor": [None],
    },
    {
        "regressor": [RandomForestRegressor(random_state=RANDOM_STATE)],
        "regressor__n_estimators": range(50, 600, 50),
        "regressor__max_depth": range(3, 8),
        "regressor__min_samples_split": range(2, 203, 25),
        "regressor__min_samples_leaf": range(25, 300, 25),
        "preprocessor": [None],
    },
]
```

```python
# список с названиями моделей
model_list = [
    "DummyRegressor",
    "LinearRegression",
    "Ridge",
    "Lasso",
    "CatBoostRegressor",
    "LGBMRegressor",
    "RandomForestRegressor",
]
```

```python
%%time

# таблица с построенными моделями
results = []

# обучение нескольких классификаторов, сохранение результатов
for params, model_name in zip(parameters, model_list):
    # создание экземпляра GriddSearchCV
    rs = RandomizedSearchCV(
                    pipe,
                    params,
                    n_iter=60,
                    cv=5,
                    verbose=1,
                    random_state=RANDOM_STATE,
                    scoring='neg_mean_absolute_error',
                    n_jobs=-1,
                    error_score='raise')
                    
    # запуск поиска по сетке
    rs.fit(train.drop(columns=['temp_last']), train['temp_last'])

    # результаты кросс-валидации
    cv_results = pd.DataFrame(rs.cv_results_)
    fit_time = cv_results[cv_results['rank_test_score'] == 1]['mean_fit_time'].values[0]
    predict_time = cv_results[cv_results['rank_test_score'] == 1]['mean_score_time'].values[0]
    dict_to_add = {'модель': model_name,
                   'mae': rs.best_score_,
                   'Время обучения': fit_time,
                   'Время предсказания': predict_time,
                   'rs': rs}
    results.append(dict_to_add)
```

```python
results_2 = pd.DataFrame(results).sort_values(by='mae', ascending=False).reset_index(drop=True)
results_2
```
